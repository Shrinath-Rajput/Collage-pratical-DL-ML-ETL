```python
#import lib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mlt

```

```python
df=pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
df.isnull()
```

```python
# independent and dependent value

x=df.iloc[: ,0].astype(str)
y=df.iloc[:,1]
```

```python
#train test split

from sklearn.model_selection import train_test_split
x_test,y_test,x_train,y_train=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
x_test
```

```python
x_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer(stop_words='english' ,max_features=3000)
X_train_vec=vec.fit_transform(x_train)
X_test_vec=vec.transform(x_test)
```

```python
print(X_test_vec.shape)
print(y_train.shape)
```

```python
#model train
from sklearn.linear_model import Perceptron
p=Perceptron()
p.fit(X_train_vec,y_train)
```
