```python
import numpy as np

# Dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Initial parameters
m = 0
b = 0

learning_rate = 0.01
iterations = 1000
n = len(x)

for i in range(iterations):
    
    y_pred = m * x + b
    
    # Cost function (Mean Squared Error)
    cost = (1/n) * sum((y_pred - y)**2)
    
    # Gradient calculation
    dm = -(2/n) * sum(x * (y - y_pred))
    db = -(2/n) * sum(y - y_pred)
    
    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db
    
    if i % 100 == 0:
        print(f"Iteration {i}: Cost={cost}, m={m}, b={b}")

print("\nFinal values:")
print("Slope (m) =", m)
print("Intercept (b) =", b)
```

```python
# Perceptron for NOT Gate

def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

# weight and bias
w = -1
b = 0.5

inputs = [0, 1]

print("NOT Gate using Perceptron")
for x in inputs:
    y = step_function(w * x + b)
    print(x, "->", y)
```
