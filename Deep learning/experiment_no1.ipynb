```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Perceptron
```

```python
df=pd.read_csv("placement.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.scatter(df['cgpa'],df['iq'],df['placement'])
```

```python
x=df[['iq','cgpa']]
y=df['placement']
```

```python
x
```

```python
y
```

```python
p=Perceptron()
```

```python
p.fit(x,y)
```

```python
print(p.intercept_)
```

```python
print(p.coef_)
```

```python
z=p.score(x,y)
print("Accuracy score is",z)
```
