```python
#missing value
print(df.isnull().sum())
```

```python
#drop with missing value
print(df.dropna)
```

```python
#fill the missing value with a specific value
mean_value=df['iq'].mean()
df['iq'].fillna(mean_value,inplace=True)
```

```python
df=pd.DataFrame(df)
```

```python
print(df)
```

```python
#Conversion to the Tensor format

import numpy as np
import tensorflow as tf
```

```python
X=df[['cgpa','placement']].values
y=df['iq'].values
```

```python
X
```

```python
y
```

```python
#convert to tensorflow tensor 
X_tensor=tf.constant(X,dtype=tf.float32)
y_tensor=tf.constant(y,dtype=tf.int32)
```

```python
#Alternatively you can use numpy array directly as Tensorflow

X_tensor=tf.convert_to_tensor(X,dtype=tf.float32)
y_tensor=tf.convert_to_tensor(y,dtype=tf.int32)
```

```python
print("\n X tensor \n",X_tensor)
```

```python
print("\n Y Tensor \n",y_tensor)
```
